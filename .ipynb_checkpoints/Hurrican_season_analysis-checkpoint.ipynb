## 🔔 Task 2 

We selected the variable `Hurricanes` as the regressor because it shows the **strongest linear relationship** with the response variable `StormIndex`. This was evident from the **correlation heatmap**, where the correlation between `Hurricanes` and `StormIndex` was the highest among all numerical predictors.


```r
# Fit the linear regression model
model_hurricanes <- lm(StormIndex ~ Hurricanes, data = ex1028)

# Show the summary
summary(model_hurricanes)
```

#### Interpretation of the coefficients:

- **Intercept (-13.27):** This is the predicted StormIndex when the number of hurricanes is zero. (However, this value is not very meaningful in practice — -13.27 is not realistic, and a hurricane season with zero hurricanes may fall outside the typical observed range.)


- **Hurricanes (19.86):** Each additional hurricane increases the expected StormIndex by approximately **19.86 units**. (the number of hurricanes has a **strong and positive effect** on the intensity of the hurricane season)

The **p-value** for `Hurricanes` is extremely low (**< 0.001**), indicating a **statistically significant relationship** between the number of hurricanes and the StormIndex.

The **R-squared value is 0.747**, meaning that about **74.7% of the variability** in `StormIndex` is explained by the number of hurricanes.

```r
# Plot data and regression line
plot(ex1028$Hurricanes, ex1028$StormIndex,
     main = "StormIndex vs. Number of Hurricanes",
     xlab = "Hurricanes", ylab = "StormIndex",
     pch = 16, col = "blue")
abline(model_hurricanes, col = "red", lwd = 2)
```

The scatter plot shows a clear **positive linear relationship** between the number of hurricanes and the StormIndex. 

---

#### Conclusion

- The number of hurricanes (`Hurricanes`) is a **strong predictor** of the storm season intensity (`StormIndex`).
- The model explains nearly **three-quarters of the variation** in the response.

This makes `Hurricanes` a very suitable regressor for modeling `StormIndex`.


## 🔔 Task 3 

We now investigate whether the type of year according to the El Niño phenomenon (`ElNino`) has an influence on the `StormIndex`. To test whether there are significant differences in `StormIndex` between the three categories of the `ElNino` variable, we apply an **analysis of variance (ANOVA)**.

```r
anova_model <- aov(StormIndex ~ ElNino, data = ex1028)

summary(anova_model)
```

- The **F-statistic** is 12.09, which suggests that the differences between group means (for different ElNino types) are relatively large.
- The **p-value** is approximately 0.00006 (very small), which indicates that these differences are **statistically significant**.

To understand how exactly the categories differ from each other, we fit a linear model and inspect the estimated coefficients.

```r
# Boxplot for visualisation
boxplot(StormIndex ~ ElNino, data = ex1028,
        col = c("lightblue", "lightgreen", "salmon"),
        main = "StormIndex by ElNino Category",
        ylab = "StormIndex", xlab = "ElNino")
```

```r
model_el <- lm(StormIndex ~ ElNino, data = ex1028)
summary(model_el)
```

#### ⚖️ Interpretation of Coefficients

- The intercept (`140.06`) represents the **mean StormIndex during "cold" El Niño years** (the baseline group).

- During **neutral years**, the average StormIndex is **about 39.62 units lower** than in cold years (statistically significant, _p_ = 0.0161).

- During **warm years**, the StormIndex is **about 77.88 units lower** than in cold years (highly significant, _p_ = 1.21e-05).

These differences are meaningful in practice: **warmer El Niño years are associated with significantly weaker hurricane seasons**, as reflected by lower `StormIndex` values.

---

####  ❓ Model Quality

- **R-squared = 0.3495**, meaning that about **35% of the variability** in `StormIndex` is explained by the El Niño category.

- This is **moderate explanatory power**, which confirms that El Niño type plays a role but there are other important factors involved.

## 🔔Task 4

We now investigate whether the number of hurricanes (Hurricanes, numeric predictor) and the type of El Niño (ElNino, categorical predictor) have a combined effect on the StormIndex. To do this, we build a linear model that includes both main effects and their interaction:

```r
model_el_hur <- lm(StormIndex ~ ElNino * Hurricanes, data = ex1028)
summary(model_el_hur)
```

### ⚖️ Interpretation of Coefficients

* **Intercept (-40.09):** This is the expected StormIndex when there are zero hurricanes and the El Niño type is cold (the baseline). Although not very meaningful in real terms (since zero hurricanes is rare), it provides a reference for interpreting the rest of the model.

* **ElNinoneutral (60.64):** When the El Niño type is neutral, and there are zero hurricanes, the StormIndex is about `60.64` units higher than in cold years).

* **ElNinowarm (44.39):** In warm El Niño years with zero hurricanes, the StormIndex is about `44.39` units higher than in cold years. This effect is marginally significant, suggesting a weaker trend.

* **Hurricanes (25.74):** For cold El Niño years, each additional hurricane increases the StormIndex by `25.74` units.

* **ElNinoneutral:Hurricanes (-12.95):** In neutral years, the impact of each hurricane is `12.95` units less than in cold years. 

* **ElNinowarm:Hurricanes (-11.26):** In warm years, each hurricane has `11.26` units less impact compared to cold years, resulting in a net effect of roughly +14.47 per hurricane.

---

### 🔩 Explore the significance of each component.

* **(Intercept)** has a p-value of `0.051`, which is just around the `0.05` threshold. This means it's marginally significant, suggesting the baseline StormIndex (for cold years with 0 hurricanes) is close to being statistically meaningful.

* **ElNinoneutral** has a p-value of `0.027`, which is below `0.05`. Therefore, it is statistically significant, indicating that neutral El Niño years lead to a significantly different StormIndex compared to cold years.

* **ElNinowarm** has a p-value of `0.094`, which is above `0.05`. This makes it not statistically significant at the `5%` level. So, there is only weak evidence that warm years differ from cold years in their effect on StormIndex.

* **Hurricanes** has a p-value of `6.2e-12`, which is extremely small. This means it is highly significant, confirming that the number of hurricanes has a strong influence on the StormIndex.

* The interaction term **ElNinoneutral:Hurricanes** has a p-value of `0.0014`, making it statistically significant at the `1%` level. This shows that the effect of hurricanes on StormIndex is moderated (reduced) in neutral El Niño years.

* Similarly, **ElNinowarm:Hurricanes** has a p-value of `0.0217`, which is below `0.05`, so it is statistically significant. This suggests that the hurricane effect is also moderated during warm El Niño years.

---

### 📈 Graphically display the regression relationship appropriately.

```r
# Generate predictions
new_data <- expand.grid(
  Hurricanes = seq(min(ex1028$Hurricanes), max(ex1028$Hurricanes), length.out = 100),
  ElNino = levels(ex1028$ElNino)
)

new_data$StormIndex_pred <- predict(model_el_hur, newdata = new_data)

options(repr.plot.width=30, repr.plot.height=10)
# Plot
ggplot(ex1028, aes(x = Hurricanes, y = StormIndex, color = ElNino)) +
  geom_point(size = 3) +
  geom_line(data = new_data, aes(y = StormIndex_pred)) +
  labs(title = "StormIndex vs Hurricanes by El Niño Type",
       x = "Number of Hurricanes", y = "StormIndex") + 
    theme(
    text = element_text(size = 14),
    plot.title = element_text(size = 24, face = "bold"),
    axis.title = element_text(size = 20),
    axis.text = element_text(size = 12)
    )
```

---

### ❓ Model Quality

* R-squared = `0.8379`, meaning the model explains approximately `84%` of the variability in StormIndex.

* This is a strong model, indicating that both El Niño type, hurricane count, and their interaction are important predictors.

* All predictors and interaction terms are statistically significant or close to significant `(p < 0.1)`, supporting their inclusion in the model.

## 🔔 Task 6

### 1. Normality of Residuals:
   * **Why**: Linear regression assumes that the residuals (model errors) are normally distributed.
   * **Test**: Shapiro-Wilk test

```r
shapiro.test(residuals(model_el_hur_afr))
```

We applied the Shapiro-Wilk test to check if the residuals are normally distributed. The test statistic was `W = 0.987` with a p-value of `0.861`. Since the p-value is high, we do not reject the null hypothesis of normality. Therefore, the normality assumption is satisfied.

### 2. Homoscedasticity (Equal Variance of Residuals)
* **Why**: Residuals should have constant variance (no systematic change in spread).
* **Test**: Breusch-Pagan test

The Breusch-Pagan test was used to verify whether the variance of the residuals is constant. The test returned `BP = 8.77` with a p-value of `0.187`. Again, the null hypothesis of constant variance is not rejected, so the homoscedasticity assumption holds.

```r
#install.packages("lmtest")
library(lmtest)
bptest(model_el_hur_afr)
```

### 3. Independence of Observations
* **Why**: Observations should be independent of each other.
* **Test**: Durbin-Watson test

```r
dwtest(model_el_hur_afr)
```

Using the Durbin-Watson test, we checked for autocorrelation in residuals. The test result was `DW = 2.02` with a p-value of `0.461`. A value close to 2 suggests no autocorrelation, so the assumption of independence is met.

### 4. Multicollinearity
* **Why**: High correlation between predictors can distort the coefficient estimates.
* **Test**: VIF – Variance Inflation Factor
* **Interpretation**: VIF > 5 (sometimes 10) indicates a potential problem.

```r
library(car)
vif(model_el_hur_afr, type = "predictor")
```

**Conclusion**: All adjusted GVIF values are close to 1, indicating that there is no evidence of problematic multicollinearity. This means our predictors do not significantly overlap in the information they contribute to the model.
